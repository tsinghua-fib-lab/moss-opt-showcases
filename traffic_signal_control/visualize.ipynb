```python
from moss import Map,Agents
import matplotlib.pyplot as plt
```

```python
# read the map
M=Map('data/hangzhou/map.bin')
# read the agents
A=Agents('data/hangzhou/agents.bin')
```

```python
# show the map

plt.subplot(1,2,1)
for lane in M.lanes:
    x,y=lane.geom.numpy().T
    plt.plot(x,y)
plt.axis('equal')

plt.subplot(1,2,2)
for lane in M.lanes:
    x,y=lane.geom.numpy().T
    plt.plot(x,y)
plt.axis('equal')
plt.xlim(348250,348350)
plt.ylim(-1093550,-1093450)

plt.gcf().set_size_inches(10,5)
plt.tight_layout()
```

```python
# show the histogram of agents' departure time

plt.hist([agent.departure_time for agent in A.agents])
```

```python
# plot agent route


# background
for road in M.roads:
    x,y=road.lanes[0].geom.numpy().T
    plt.plot(x,y,'gray',zorder=0)
# the route of agent 0
route=A.agents[0].route
for road in route:
    x,y=M.road_map[road].lanes[0].geom.numpy().T
    plt.plot(x,y,'red',zorder=1)
```
